```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#Leitura dos arquivos Importados diretamente aqui
df1 = pd.read_excel('Base Vendas.xlsx')

```

Calculos

```python
# Calculos
#CUSTO do PRODUTO
df1['Custo'] = df1['Custo Unitario'].mul(df1['Tamanho Pedido'])
#Valor da Venda
df1['Preco'] = df1['Preco Unitario'].mul(df1['Tamanho Pedido'])
#CALCULANDO O LUCRO
df1['Lucro'] = df1['Preco'] - df1['Custo']

```

Graficos VENDAS

```python
#GRAFICO TOTAL DE PRODUTOS VENDIDOS
df1.groupby('Marca')['Tamanho Pedido'].sum().sort_values(ascending=True).plot.barh(title='Total de vendas por Marca' )
plt.xlabel('Total')
plt.ylabel('Produto')
```

```python
# LUCRO POR ANO
df1.groupby(df1['Data da Venda'].dt.year)['Lucro'].sum().plot.bar(title='Lucro por ano')
plt.xlabel('ano')
plt.ylabel('receita')
```

```python
# LUCRO POR Loja
df1.groupby('Loja')['Tamanho Pedido'].sum().sort_values(ascending=True).plot.barh(title='Total de vendas por Loja' )
plt.xlabel('Total')
plt.ylabel('Loja')
```

```python
#LUCRO POR LOJA
df1.groupby('Loja')['Lucro'].sum().sort_values(ascending=True).plot.bar(title='Lucro x Loja' )
plt.xlabel('Total')
plt.ylabel('Loja')
```

```python
#SELECIONANDOA APENAS VENDAS DE 2018
df_2018 = df1[df1['Data da Venda'].dt.year == 2018]
```

```python
df_2018.groupby(df_2018['Data da Venda'].dt.month)['Lucro'].sum().plot(title='Lucro x Mes')
plt.xlabel('mes')
plt.ylabel('Lucro')
```

```python
df_2018.groupby(df_2018['Loja'])['Lucro'].sum().plot(title='Lucro x LOJA 2018')
plt.xlabel('Loja')
plt.ylabel('Lucro')
```
